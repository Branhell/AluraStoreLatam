# Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Cálculo del ingreso total por tienda
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

# Mostrar resultados
print("Ingreso total Tienda 1:", ingreso_tienda1)
print("Ingreso total Tienda 2:", ingreso_tienda2)
print("Ingreso total Tienda 3:", ingreso_tienda3)
print("Ingreso total Tienda 4:", ingreso_tienda4)
```

```python
import matplotlib.pyplot as plt

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, ingresos, color='skyblue')
plt.title("Ingreso Total por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingreso Total")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
ventas_categoria1 = tienda["Categoría del Producto"].value_counts()
ventas_categoria2 = tienda2["Categoría del Producto"].value_counts()
ventas_categoria3 = tienda3["Categoría del Producto"].value_counts()
ventas_categoria4 = tienda4["Categoría del Producto"].value_counts()

print("Ventas por categoría - Tienda 1:\n", ventas_categoria1)
print("\nVentas por categoría - Tienda 2:\n", ventas_categoria2)
print("\nVentas por categoría - Tienda 3:\n", ventas_categoria3)
print("\nVentas por categoría - Tienda 4:\n", ventas_categoria4)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Ventas por Categoría en Cada Tienda", fontsize=16)

axs[0, 0].bar(ventas_categoria1.index, ventas_categoria1.values, color='lightcoral')
axs[0, 0].set_title("Tienda 1")
axs[0, 0].tick_params(axis='x', rotation=45)

axs[0, 1].bar(ventas_categoria2.index, ventas_categoria2.values, color='lightblue')
axs[0, 1].set_title("Tienda 2")
axs[0, 1].tick_params(axis='x', rotation=45)

axs[1, 0].bar(ventas_categoria3.index, ventas_categoria3.values, color='lightgreen')
axs[1, 0].set_title("Tienda 3")
axs[1, 0].tick_params(axis='x', rotation=45)

axs[1, 1].bar(ventas_categoria4.index, ventas_categoria4.values, color='gold')
axs[1, 1].set_title("Tienda 4")
axs[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# 3. Calificación promedio de la tienda


```python
valoracion_media1 = tienda["Calificación"].mean()
valoracion_media2 = tienda2["Calificación"].mean()
valoracion_media3 = tienda3["Calificación"].mean()
valoracion_media4 = tienda4["Calificación"].mean()

print("Valoración media Tienda 1:", round(valoracion_media1, 2))
print("Valoración media Tienda 2:", round(valoracion_media2, 2))
print("Valoración media Tienda 3:", round(valoracion_media3, 2))
print("Valoración media Tienda 4:", round(valoracion_media4, 2))
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
valoraciones = [valoracion_media1, valoracion_media2, valoracion_media3, valoracion_media4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, valoraciones, color='mediumseagreen')
plt.title("Valoración Media por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación Promedio")
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
productos_tienda1 = tienda["Producto"].value_counts()
productos_tienda2 = tienda2["Producto"].value_counts()
productos_tienda3 = tienda3["Producto"].value_counts()
productos_tienda4 = tienda4["Producto"].value_counts()

print("Tienda 1 - Más vendido:\n", productos_tienda1.head(1))
print("Tienda 1 - Menos vendido:\n", productos_tienda1.tail(1))

print("\nTienda 2 - Más vendido:\n", productos_tienda2.head(1))
print("Tienda 2 - Menos vendido:\n", productos_tienda2.tail(1))

print("\nTienda 3 - Más vendido:\n", productos_tienda3.head(1))
print("Tienda 3 - Menos vendido:\n", productos_tienda3.tail(1))

print("\nTienda 4 - Más vendido:\n", productos_tienda4.head(1))
print("Tienda 4 - Menos vendido:\n", productos_tienda4.tail(1))
```

```python
def graficar_top5(productos, titulo, color):
    top5 = productos.head(5)
    plt.bar(top5.index, top5.values, color=color)
    plt.title(titulo)
    plt.xticks(rotation=45)
    plt.ylabel("Cantidad Vendida")
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
graficar_top5(productos_tienda1, "Top 5 Productos - Tienda 1", 'lightcoral')

plt.subplot(2, 2, 2)
graficar_top5(productos_tienda2, "Top 5 Productos - Tienda 2", 'lightblue')

plt.subplot(2, 2, 3)
graficar_top5(productos_tienda3, "Top 5 Productos - Tienda 3", 'lightgreen')

plt.subplot(2, 2, 4)
graficar_top5(productos_tienda4, "Top 5 Productos - Tienda 4", 'gold')

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
envio_promedio1 = tienda["Costo de envío"].mean()
envio_promedio2 = tienda2["Costo de envío"].mean()
envio_promedio3 = tienda3["Costo de envío"].mean()
envio_promedio4 = tienda4["Costo de envío"].mean()

print("Envío promedio Tienda 1:", round(envio_promedio1, 2))
print("Envío promedio Tienda 2:", round(envio_promedio2, 2))
print("Envío promedio Tienda 3:", round(envio_promedio3, 2))
print("Envío promedio Tienda 4:", round(envio_promedio4, 2))
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
envios = [envio_promedio1, envio_promedio2, envio_promedio3, envio_promedio4]

plt.figure(figsize=(8,5))
plt.bar(tiendas, envios, color='orange')
plt.title("Costo de Envío Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo Promedio de Envío")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# **Informe Final – Análisis de Tiendas Alura Store**

## I Introducción

El presente informe tiene como objetivo analizar el desempeño de las cuatro tiendas de la cadena Alura Store con el fin de recomendar al Sr. Juan cuál de ellas debería vender para iniciar un nuevo emprendimiento. Se evaluaron métricas clave como ingresos, categorías más vendidas, satisfacción del cliente, productos destacados y costos de envío. Los resultados se presentan a continuación, acompañados de visualizaciones que permiten interpretar los datos de forma clara y efectiva.

## II Desarrollo del Análisis

### II.I Ingresos Totales

Se calculó el ingreso total de cada tienda sumando los valores de la columna Precio. Esta métrica permite identificar qué tienda genera más ingresos y cuál tiene menor rendimiento económico.

### II.II Categorías Más Vendidas

Se agruparon los productos por categoría para conocer cuáles son las más populares en cada tienda. Esto ayuda a entender el enfoque comercial y la diversidad de productos ofrecidos.

### II.III Calificación Promedio

Se calculó la media de las calificaciones de los clientes, lo que refleja el nivel de satisfacción con los productos adquiridos en cada tienda.

### II.IV Productos Más y Menos Vendidos

Se identificaron los productos con mayor y menor frecuencia de venta, lo que permite detectar tendencias y oportunidades de mejora en el catálogo.

### II.V Costo de Envío Promedio

Se analizó el costo de envío promedio por tienda, considerando que este valor es asumido por el cliente. Un costo elevado puede afectar la percepción de valor y la decisión de compra.

### II.VI Visualizaciones

Se generaron gráficos de barras, circulares y de dispersión para complementar el análisis y resaltar los patrones más relevantes:

* Ingreso total por tienda.

* Distribución de categorías.

* Relación entre precio y calificación.
